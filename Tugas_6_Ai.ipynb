```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# 1. Generate spatial-like data
X, y = make_moons(n_samples=300, noise=0.1, random_state=42)

# 2. Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. DBSCAN clustering
db = DBSCAN(eps=0.3, min_samples=5)
db.fit(X_scaled)

# 4. Hasil clustering
labels = db.labels_

# Label -1 menunjukkan outlier
core_samples_mask = (labels != -1)
outliers_mask = (labels == -1)

# 5. Visualisasi
plt.figure(figsize=(8, 6))
plt.scatter(X_scaled[core_samples_mask, 0], X_scaled[core_samples_mask, 1], c=labels[core_samples_mask], cmap='viridis', label='Clustered Points')
plt.scatter(X_scaled[outliers_mask, 0], X_scaled[outliers_mask, 1], c='red', marker='x', label='Outliers')
plt.title("DBSCAN Clustering with Anomaly Detection (make_moons data)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
